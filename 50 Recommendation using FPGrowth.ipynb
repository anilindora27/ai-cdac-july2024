##### Create the dataset

```python
dataset = [
    ['Apple','Beer','Rice','Chicken'],
    ['Apple','Beer','Rice'],
    ['Apple','Beer'],
    ['Apple','Peer'],
    ['Milk','Beer','Rice','Chicken'],
    ['Milk','Beer','Rice'],
    ['Milk','Beer'],
    ['Milk','Peer'],
]
```

```python
dataset
```

```python
# above data is unstructured data , we need to convert it structured form 
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded = te.fit_transform(dataset)
```

```python
encoded
```

```python
te.columns_
```

```python
df = pd.DataFrame(encoded, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
# 1 Find frequent itemset
freq_itemset = fpgrowth(df, min_support=0.25, use_colnames= True)
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.5)
```

```python
rules.shape
```

```python
rules
```

```python
# extracting four columsn only
rules = rules[['antecedents','consequents','support','confidence']]
```

```python
rules
```

##### Sort the rules

```python
# by confidence , descending order
```

```python
rules.sort_values(by=['confidence','support'], ascending = False)
```

##### Filter the rules

```python
nrules = rules[rules['confidence'] >0.5]
```

```python
nrules.shape
```

```python
nrules
```

##### Create Recommendation

```python
nrules[nrules['antecedents']=={'Rice'}]
```

```python
nrules[nrules['antecedents']=={'Rice'}]['consequents']
```

```python
# if two are there 
nrules[nrules['antecedents']=={'Apple','Beer'}]['consequents']
```

##### Export the rules

```python
nrules.to_csv('rules_fpgrowth.csv', index = False)
```
