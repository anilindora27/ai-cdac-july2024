```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

##### Create the dataset

```python
# Step 1: Create a sample dataset
x,y = make_blobs(n_samples=300, centers= 4, cluster_std= 0.60, random_state=0) # no of columns by default 2
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2')
plt.scatter(x[:,0], x[:,1])
```

##### Create the object

```python
dbscan = DBSCAN(eps = 0.6, min_samples= 5)
clusters = dbscan.fit_predict(x)
```

```python
clusters
```

```python
##### Visualize the cluster
```

```python
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2')
plt.scatter(x[:,0], x[:,1], c = clusters)
```

```python
# it create the clusters by its own, there is no need to give value of k here
```
