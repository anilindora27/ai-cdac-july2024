```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/lenna.png')
```

```python
# Define transformation parameters
tx = 50 # Translation along x axis
ty = 100 # Translation along y axis
angle = 45 # Rotation angle in degree
scale = 0.75 # Scaling factor
shear_x = 0.3 # Shear along x asis
shear_y = 0.2 # Shear along y axis

# Create transformation matrix
M = np.array([[scale*np.cos(np.radians(angle)), shear_x, tx],
              [shear_y, scale*np.sin(np.radians(angle)), ty]])

# Apply transformationn using warpAffine
img_affine = cv2.warpAffine(image, M, (512,512))

#  Display original and affine transformed images
plt.subplot(1,2,1)
plt.imshow( image)
plt.subplot(1,2,2)
plt.imshow( img_affine)
```

```python
final_img = np.hstack([image, img_affine])
```
