```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('images/lenna.png',0)
```

```python
# Sobel parameters: ksize , dx, dy, depth
# Sobel for horizontal edges
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize =3)
 # Sobel for vertical edges
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize =3)

# Combine the Sobel images 
combined_sobel = cv2.add(sobelx,sobely)

```

```python
final_img =np.hstack([sobelx,sobely,combined_sobel])
cv2.imshow('All Images', final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141) # We can also use like this without commas
plt.title('Original image')
plt.imshow(img, cmap ='gray')

plt.subplot(142)
plt.title(' sobel X Image')
plt.imshow(abs(sobelx), cmap = 'gray')

plt.subplot(143)
plt.title('  sobel Y Image')
plt.imshow(abs(sobely), cmap = 'gray')

plt.subplot(144)
plt.title('  X + Y Image')
plt.imshow(abs(combined_sobel), cmap = 'gray')
```

##### Canny Edge Detection

```python
img = cv2.imread('images/lenna.png',0)
```

```python
# Apply Canmy edge detection
edges = cv2.Canny(img, 50 , 150) # adjust threshold if neede

# display results
cv2.imshow('Original Image', img)
cv2.imshow('Canny Edge Detection', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
