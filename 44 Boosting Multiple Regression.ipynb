```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/mtcars.csv')
```

```python
df
```

```python
# mpg (mileage is output ) 
# for now taking input only 3 # disp , hp, wt
```

```python
# disp, hp, wt
x = df[['disp','hp','wt']]

# mpg
y = df['mpg']

```

```python
x.corrwith(y)
```

```python
# -ve show that inverse proportion
```

##### Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
# r2 score
regressor.score(x,y) # accuracy of the model
```

```python
regressor.coef_ # negatively correlated , as values are negative , -ve connected
```

```python
regressor.intercept_
```

##### Predict on unseen data

```python
disp = 221
hp = 102
wt = 2.91
```

```python
new = pd.DataFrame({
    'disp': [221],
    'hp': [102],
    'wt' : [2.91]
})
```

```python
new
```

```python
regressor.predict(new)
```

```python
#regressor.predict([[221,102,2.91]])
```

```python
x
```

```python
disp * regressor.coef_[0] + hp * regressor.coef_[1] + wt * regressor.coef_[2] + regressor.intercept_
```

##### Visualize

```python
plt.figure(figsize=(16,5))
plt.subplot(131)
plt.title('disp vs mpg')
plt.xlabel('disp')
plt.ylabel('mpg')
plt.scatter(x['disp'], y)
plt.grid()

plt.subplot(132)
plt.title('hp vs mpg')
plt.xlabel('hp')
plt.ylabel('mpg')
plt.scatter(x['hp'], y)
plt.grid()

plt.subplot(133)
plt.title('wt vs mpg')
plt.xlabel('wt') 
plt.ylabel('mpg')
plt.scatter(x['wt'], y)
plt.grid()
```

##### AdaBoost the regression

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
reg_adb= AdaBoostRegressor(regressor, n_estimators=10, random_state=0)
```

```python
reg_adb.fit(x, y)
```

##### Performance Evaluation after AdaBoosting

```python
reg_adb.score(x,y) # increased by 0.5 % only after boosting
```
