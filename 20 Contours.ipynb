```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('images/houseleft.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create a binary image pixels above 127 are set 255 (white),
# and others to 0 (black)
thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)[1]

# cv2.findContours finds contours in the binary image.
# cv2.RETR_TREE retrives all contours and creates a hierarcy tree.
# cv2.CHAIN_APPROX_SIMPLE approximates contours to save memory.
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# -1 draws all contours.
# (0,255, 0) is the color (green)
# 2 is the thicknses of the lines.
cv2.drawContours(gray, contours, -1, (0,255,0), 2)
# Draw all contours in green 

# Display results
cv2.imshow('Original Image', gray)
cv2.imshow('Threshold Image', thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121) # We can also use like this without commas
plt.title('Original image')
plt.imshow(gray)

plt.subplot(122)
plt.title(' Edges Detected')
plt.imshow(thresh, cmap = 'gray' )
```
