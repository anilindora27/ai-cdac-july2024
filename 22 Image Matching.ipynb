```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image1 = cv2.imread('images/eiphel1.png',0)
image2 = cv2.imread('images/eiphel2.png',0)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(121) # We can also use like this without commas
plt.title('Image 1')
plt.imshow(image1,cmap = 'gray' )

plt.subplot(122)
plt.title(' Image 2 ')
plt.imshow(image2, cmap = 'gray' )
```

```python
sift = cv2.SIFT_create()
```

```python
keypoints_1, description_1 = sift.detectAndCompute(image1, None)
keypoints_2, description_2 = sift.detectAndCompute(image2, None)

len(keypoints_1), len(keypoints_2)
```

```python
# feature matching
bf = cv2.BFMatcher(cv2.NORM_L1)

matches = bf.match(description_1, description_2)
mathes = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(image1, keypoints_1, image2, keypoints_2, matches[:40], image2, flags= 2)
plt.imshow(img3)
```
