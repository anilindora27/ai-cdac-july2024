```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
new_img = img > 150
inv_img = img <150
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131) # We can also use like this without commas
plt.title('Original image')
plt.imshow(img, cmap ='gray')

plt.subplot(132)
plt.title(' Binary Image')
plt.imshow(new_img, cmap = 'gray')

plt.subplot(133)
plt.title(' Inverse Binary Image')
plt.imshow(inv_img, cmap = 'gray')
```

```python
plt.figure(figsize=(16,9))
plt.title('Binary Image Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(0,256,10))
plt.grid()
plt.hist(new_img.flatten(), bins= range(0, 256, 10));
```

##### Otsu Threshold

```python
# Apply Otsu's threholding
thresh_val, thresh_otsu = cv2.threshold(img, 0,255, cv2.THRESH_OTSU)
```

```python
thresh_val
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141) # We can also use like this without commas
plt.title('Original image')
plt.imshow(img, cmap ='gray')

plt.subplot(142)
plt.title(' Binary Image')
plt.imshow(new_img, cmap = 'gray')

plt.subplot(143)
plt.title(' Inverse Binary Image')
plt.imshow(inv_img, cmap = 'gray')

plt.subplot(144)
plt.title('  Otsu Threshold Image')
plt.imshow(thresh_otsu, cmap = 'gray')
```

```python
thresh_val, thresh_mask = cv2.threshold(img, 0,255, cv2.THRESH_MASK)

plt.figure(figsize=(16,9))
plt.subplot(151) # We can also use like this without commas
plt.title('Original image')
plt.imshow(img, cmap ='gray')

plt.subplot(152)
plt.title(' Binary Image')
plt.imshow(new_img, cmap = 'gray')

plt.subplot(153)
plt.title(' Inverse Binary Image')
plt.imshow(inv_img, cmap = 'gray')

plt.subplot(154)
plt.title('  Otsu Threshold Image')
plt.imshow(thresh_otsu, cmap = 'gray')

plt.subplot(154)
plt.title('  Mask Threshold Image')
plt.imshow(thresh_mask, cmap = 'gray')
```
