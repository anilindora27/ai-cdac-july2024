##### Polynomial Regression

```python
df = pd.read_csv('datasets/Salary_Data.csv')
x = df[['YearsExperience']]
y = df['Salary']
```

```python
#  Using  Simple Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
print('Accuracy:', reg.score(x, y))
```

```python
# after polynomial transformation , we will see the accuracy 
poly = PolynomialFeatures( degree = 2 )
x_poly = poly.fit_transform(x)
```

```python
x_poly
```

```python
reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y ))
```

```python
# now change degree to 4, it will increase the accuracy of the model
poly = PolynomialFeatures( degree = 4 )
x_poly = poly.fit_transform(x)
```

```python
reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y ))
```

```python
# keep changing the degree and see how the accuracy changes
poly = PolynomialFeatures( degree = 10)
x_poly = poly.fit_transform(x)

reg.fit(x_poly, y)
print('Accuracy:', reg.score(x_poly, y ))
```

```python
acc = []
for i in range(2,31):
    poly = PolynomialFeatures( degree = i)
    x_poly = poly.fit_transform(x)
    reg.fit(x_poly, y)
    acc.append(reg.score(x_poly,y))
acc
```

```python
plt.grid()
plt.xticks(range (2,31))
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.plot(range(2,31), acc, color = 'r', lw =2, marker = 'o')
```

```python
poly = PolynomialFeatures( degree = 12)
x_poly = poly.fit_transform(x)

reg.fit(x_poly, y)
reg.score(x_poly, y )
```

```python
y_pred = reg.predict(x_poly)
```

```python
plt.grid()
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(x,y, color = 'green' , label = 'actual')
plt.plot(x, y_pred, color = 'red', label = 'predicted')
plt.legend()
```
