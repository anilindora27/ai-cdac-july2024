```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num = [np.random.randint(0, 100) for x in range(1000)]
```

```python
num
```

```python
print(num)
```

```python
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(0,101,10))
plt.grid()
plt.hist(num, bins= range(0, 101,10));
```

##### histogram for image processing

```python
import cv2
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
plt.figure(figsize=(16,9))
plt.title('Image Histogram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(0,256,10))
plt.grid()
plt.hist(img.flatten(), bins= range(0, 256, 10));
```

```python
# if values more tends to 255 it means image is bright
# and if tends to 0 means dark image
```
